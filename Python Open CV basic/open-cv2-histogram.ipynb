# how to show image histogram

histogram as a graph or plot which gives an overall idea about the intensity distribution of an image. it is a plot with pixel values ranging from 0 to 255, not always in x-axis and corresponding number of pixels in the image on y-axis.

```python
def watchimageShow(imageTitle,image):
    if len(image.shape)==3:
        imageVariable=cv2.cvtColor(image,cv2.COLOR_BAYER_BGR2RGB)
        if image.shape[2]==1:
            imageVariable=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    plt.figure(figsize=(4,2))
    plt.imshow(imageVariable)
    plt.title(imageTitle)
    plt.show()
```

```python
def imageHistogram(image, imageTitle,mask=None):
    color_Channel = cv2.split(image)
    color_com=('b','g','r')
    plt.figure(figsize=(4,2))
    plt.title(imageTitle)
    plt.xlabel("Bins")
    plt.ylabel("pixel Information")
    
    for (color_channel, color_comp) in zip(color_channel, color_comp):
        histogram=cv2.calcHist([color_channel],[0],mask,[256],[0,256])
        plt.plot(histogram, color=color_comp)
        plt.xlim([0,256])
        plt.show()      
    
```

```python
image=cv2.imread("../input/watchfile/watch.jpg")
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
blurImage = cv2.GaussianBlur(gray,(5,5),0)
watchimageShow("original Image", image)
watchimageShow("blur image", blurImage)
```

```python
cannyBlurWide = cv2.Canny(blurImage, 20, 250)
cannyBlurMiddle = cv2.Canny(blurImage, 15, 120)
cannyBlurFit = cv2.Canny(blurImage, 240, 250)

watchimageShow("cannyBlurWide",cannyBlurWide)
watchimageShow("cannyBlurMiddle", cannyBlurMiddle)
watchimageShow("cannyBlurFit", cannyBlurFit)
```
